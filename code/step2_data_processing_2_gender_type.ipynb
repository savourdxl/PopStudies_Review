```python
# part 1: gender
# unable to use gender_detector

def find_gender(i_article):
    
    author_info = eval(info_article.iloc[i_article]['author_info'])
    detect = gender.Detector()
    
    if author_info == []:
        return []
    
    author_info_new = []
    
    for i_author_info in author_info:
        
        name = i_author_info[0].lstrip(' ')
        
        if name == '':
            gender_author = ''
            tuple = (i_author_info[0], gender_author, i_author_info[1], i_author_info[2])
            author_info_new.append(tuple)
            continue
        else:
            name = name.strip(' ')
            name = name.lstrip('Dr. ')
            name = name.lstrip('Dr ')
            name = name.split(' ')[0]
            
            gender_author = detect.get_gender(name)
            tuple = (i_author_info[0].strip(' '), gender_author, i_author_info[1], i_author_info[2])
            author_info_new.append(tuple)
            
    print('get gender No.'+str(i_article)+' finished')
    return author_info_new 
         
```

```python
# find_gender(0)
# detect.get_gender('Thomas')
for i_article in range(info_article.shape[0]):
    
    info_article.loc[i_article, 'author_info'] = str(find_gender(i_article))
```

```python
def decide_type(i_article):
    # article   book review    
    # editorial board    editorial
    # introcution    correction
    
    try:
        type_indication_1 = info_article.iloc[i_article]['type_1'].lower().strip(' ')
    except:
        type_indication_1 = ''
    
    type_indication_2 = info_article.iloc[i_article]['type_2'].lower().strip(' ')
    type_indication_3 = info_article.iloc[i_article]['name_book'].lower().strip(' ')
    
    if type_indication_2 == 'introduction':
        my_type = 'introduction'
        
    elif type_indication_1 == 'correction':
        my_type = 'correction'
        
    elif type_indication_1 == 'book review':
        # consider "book review essays" as book reviews
        my_type = 'book review'
        
    elif type_indication_3 == 'editorial board':
        my_type = 'editorial board'
        
    elif type_indication_1 == 'review':
        # only one: volume 21 issue 2  article 8
        my_type = 'article'
        
    elif type_indication_2 == 'editorial':
        my_type = 'editorial'
        
    elif type_indication_2 == 'editorials':
        my_type = 'editorial'
    
    elif type_indication_2 == 'articles' :
        my_type = 'article'
    
    elif type_indication_2 == 'original articles':
        my_type = 'article'
    
    elif type_indication_3 == 'corrigendum':
        my_type = 'correction'
    
    elif type_indication_3 == 'erratum':
        my_type = 'correction'
        
    elif type_indication_3 == 'introduction':
        my_type = 'introduction'
        
    elif type_indication_3 == 'foreword':
        my_type = 'introduction'
    
    elif type_indication_2 == 'miscellany':
        my_type = 'article'
        
    elif type_indication_1 == 'editorial':
        my_type = 'editorial'
        
    elif 'review' in type_indication_2:
        my_type = 'book review'
      
    # put "index" into the category "article"
    else:
        my_type = 'article'
        
    print('Decision for type of Article No.'+str(i_article)+ ' done.')
    
    return my_type
```

```python
for i_article in range(info_article.shape[0]):
    
     info_article.loc[i_article, 'type_decide'] = decide_type(i_article)
```
