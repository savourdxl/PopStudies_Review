```python
import os
import pandas as pd
import numpy as np

import requests
import urllib.request
from bs4 import BeautifulSoup

from selenium import webdriver
import time

import re

path = os.path.abspath(os.path.dirname(os.getcwd())) + '\\data'
article_df = pd.read_csv(path+'\\website_article\\article.csv')
```

```python
# part 0 : information 0
info_0 = article_df.iloc[:,0:4]
info_0.to_csv(path+'\\info\\info_0.csv',index=False,sep=',')
info_0.tail()
```

```python
# part 1 : information 1

def get_info(i):
    
    url_article = article_df["link"].iloc[i]
    response = requests.get(url=url_article)
    
    web_content = response.content
    bs_obj=BeautifulSoup(web_content,'lxml').body
    
    info_1 = pd.DataFrame(columns=["type_2", "page_start","page_end","date_receive",
                                   "date_accept","date_online","keyword"])
    
    # type_2
    type_2 = bs_obj.find_all(name='div', attrs={"class":"toc-heading"})[0].h3.string.strip('\n')
    
    # page
    try:
        page_str = bs_obj.find_all(name='span', attrs={"class":"contentItemPageRange"})[0].string
        
        page_start = re.findall(r"Pages ([A-Za-z0-9]*)-", page_str)[0]
        page_end = re.findall(r"-([A-Za-z0-9]*)\n", page_str)[0]
        
    except:
        try:
            page_start = re.findall(r"Page ([A-Za-z0-9]*)\n", page_str)[0]
            page_end = page_start
        except:
            page_start = np.nan
            page_end = np.nan
    
    # date
    date_str = str(bs_obj.find_all(name='div', attrs={"class":"inline-dropzone"})[2])
    
    try:
        date_receive = re.findall(r"<div>Received ([0-9][0-9] [A-Z][a-z][a-z] [0-9][0-9][0-9][0-9])</div>", date_str)[0]
    except:
        date_receive = np.nan
    
    try:
        date_accept = re.findall(r"<div>Accepted ([0-9][0-9] [A-Z][a-z][a-z] [0-9][0-9][0-9][0-9])</div>", date_str)[0]
    except:
        date_accept = np.nan
        
    try:
        date_online = re.findall(r"<div>Published online: ([0-9][0-9] [A-Z][a-z][a-z] [0-9][0-9][0-9][0-9])</div>", date_str)[0]
    except:
        date_online = np.nan
    
    # keyword
    try:
        keyword = []
        keyword_con = bs_obj.find_all(name='div', attrs={"class":"hlFld-KeywordText"})
        
        for i_keyword_con in keyword_con:
            for i_keyword in i_keyword_con.find_all(name='a',attrs={"class":"kwd-btn keyword-click"}):
                keyword.append(i_keyword.string)
    
    except:
        keyword = []
    
    # add information 
    info_1.loc[i] = [type_2, page_start, page_end, date_receive, date_accept, date_online, keyword]
    
    return info_1
```

```python
info_1_all = pd.DataFrame(columns=["page_start","page_end","date_receive",
                                   "date_accept","date_online","keyword"])

for i in range(article_df.shape[0]):
    
    info_1 = get_info(i)
    
    info_1_all = pd.concat([info_1_all, info_1], axis=0, ignore_index=True)
    print("Getting information of article No."+str(i+1), " finished.")
```

```python
info_1_all.to_csv(path+'\\info\\info_1_o.csv',index=False,sep=',')
info_1_all.tail()
```

```python
# put information 0 and information 1 together (new information 1)
info_1_all = info_1_all[["type_2","page_start","page_end","date_receive","date_accept","date_online","keyword"]]
info_1_0 = pd.concat([info_0, info_1_all], axis=1)
info_1_0.to_csv(path+'\\info\\info_1.csv',index=False,sep=',')
info_1_0.tail()
```

```python
# part 2: information 2

def get_info_2(i):
    
    url_article = article_df["link"].iloc[i]
    response = requests.get(url=url_article)
    
    web_content = response.content
    bs_obj=BeautifulSoup(web_content,'lxml').body
    
    info_2 = pd.DataFrame(columns=["author_number","author_info"])
    
    # author
    
    try:
        author_list = bs_obj.find_all(name='span', attrs={"class":"NLM_contrib-group"})[0]
        author_list = author_list.find_all(name='a', attrs={"class":"entryAuthor"})
        
        author_number = 0
        author_info = []
        
        for i_author in author_list:
            
            i_author = str(i_author)
            
            try:
                author_name = re.findall(r">(.+?)<span class=\"overlay\">", i_author)[0]
            except:
                try:
                    author_name = re.findall(r">(.+?)</a>", i_author)[0]
                except:
                    author_name = ''
                
            try:
                author_institute = re.findall(r"<span class=\"overlay\">(.+?)<", i_author)[0]
            except:
                author_institute = ''
                
            try:
                author_href = 'https://www.tandfonline.com' + re.findall(r"<a class=\"entryAuthor\" href=\"(.+?)\">", i_author)[0]
            except:
                author_href = ''
                
            author_tuple = (author_name, author_institute, author_href)
            author_info.append(author_tuple)
            
            author_number = author_number + 1
        
    except:
        author_number = 0
        author_info = []
        
    info_2.loc[i] = [author_number, author_info]
    
    return info_2
```

```python
info_2_all = pd.DataFrame(columns=["author_number","author_info"])

for i in range(article_df.shape[0]):
    
    info_2 = get_info_2(i)
    
    info_2_all = pd.concat([info_2_all, info_2], axis=0, ignore_index=True)
    print("Getting information of article No."+str(i+1), " finished.")
```

```python
info_2_all.to_csv(path+'\\info\\info_2.csv',index=False,sep=',', encoding='utf_8_sig')
info_2_all.tail()
```

```python
# part 3 : information 3

def get_info_3(i):
    
    url_metrics = article_df["link"].iloc[i]
    
    driver = webdriver.Chrome()
    driver.get(url_metrics)
    #driver.find_elements_by_link_text('Metrics')[0].click()
    web_content = driver.page_source
    time.sleep(0.5)
    driver.quit()
    
    bs_obj=BeautifulSoup(web_content,'lxml').body
    
    # Metrics
    info_3 = pd.DataFrame(columns=["views","Crossref", "Web_of_Science"])
    
    metrics_list = bs_obj.find_all(name='div', attrs={"class":"value"})
    
    metrics = bs_obj.find_all(name='div', attrs={"class":"articleMetricsContainer"})[0]
   
    views =  metrics_list[0].string.split('\n')[1]
    Crossref = metrics_list[1].string.split('\n')[1]
    Web_of_Science = metrics_list[2].span.string
    
    info_3.loc[i] = [views, Crossref, Web_of_Science]
    
    return info_3
```

```python
info_3_all = pd.DataFrame(columns=["views","Crossref", "Web_of_Science"])

for i in range(article_df.shape[0]):
    
    info_3 = get_info_3(i)
    
    info_3_all = pd.concat([info_3_all, info_3], axis=0, ignore_index=True)
    print("Getting information of article No."+str(i+1), " finished.")
```

```python
info_3_all.to_csv(path+'\\info\\info_3.csv',index=False,sep=',', encoding='utf_8_sig')
info_3_all.tail()
```

```python
# part 4 : information 4

def get_info_4(i):
    
    url_article = article_df["link"].iloc[i]
    response = requests.get(url=url_article)
    
    web_content = response.content
    bs_obj=BeautifulSoup(web_content,'lxml').body
    
    info_4 = pd.DataFrame(columns=["name_book"])
    
    name_book_str = bs_obj.find_all(name='span', attrs={"class":"NLM_article-title hlFld-title"})[0]
    name_book = re.findall(r"\"NLM_article-title hlFld-title\">(.+?)</span>", str(name_book_str))[0]
    
    info_4.loc[i] = [name_book]
    
    return info_4
```

```python
info_4_all = pd.DataFrame(columns=["name_book"])

for i in range(article_df.shape[0]):
    
    info_4 = get_info_4(i)
    
    info_4_all = pd.concat([info_4_all, info_4], axis=0, ignore_index=True)
    print("Getting information of article No."+str(i+1), " finished.")
```

```python
info_4_all.to_csv(path+'\\info\\info_4.csv',index=False,sep=',', encoding='utf_8_sig')
info_4_all.tail()
```
