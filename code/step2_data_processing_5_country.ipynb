```python
def count_country(i_article):
    
    author_info = eval(info_article.iloc[i_article]['author_info'])
    
    country_list = []
    author_multiple_count = 0
    
    for i_author_info in author_info:
        
        try:
            countries = i_author_info[2]
            countries = eval(countries)
            
            if len(countries) > 1 :
                author_multiple_count += 1
                
        except:
            countries = []
        
        country_list = country_list + countries
        
        
    country_dict = {}
        
    for key in country_list:
        country_dict[key] = country_dict.get(key, 0) + 1
    
    country_tuple_list = []
    
    for key in country_dict:
        country_name = key
        country_number = country_dict[key]
        country_tuple = (country_name, country_number)
        
        country_tuple_list.append(country_tuple)
    
    
    return country_tuple_list, author_multiple_count
    
```

```python
def get_country_frame(info_article_k):
    
    country_list_name = []
    
    for i_article in range(info_article_k.shape[0]):
        
        country_list = eval(info_article_k.iloc[i_article]['country_list'])
        
        for i_country in country_list:
            
            i_country_name = i_country[0]
            country_list_name.append(i_country_name)
            
    country_dict = {}
    
    for key in country_list_name:
        country_dict[key] = country_dict.get(key, 0) + 1
        
    country_frame = pd.DataFrame(columns=['country_name', 'country_appear'])    
    
    for key in country_dict:
        country_name = key
        country_appear = country_dict[key]
        
        country_frame_add = pd.DataFrame([[country_name, country_appear]],
                                         columns=['country_name', 'country_appear'])
        country_frame=pd.concat([country_frame,country_frame_add])
    
    country_frame = country_frame.reset_index(drop = True)
    
    for i_country in range(country_frame.shape[0]):
        
        country_name = country_frame.iloc[i_country]['country_name']
        if country_name == 'Kosovo':
            country = '-99'
        else:
            country = pycountry.countries.get(name=country_name)
            country = country.alpha_3
            
        country_frame.loc[i_country, 'iso_a3'] = country    
    
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    world = world[world.name != "Antarctica"]
    
    world_value = pd.merge(world, country_frame, how='left')
    
    world_value.loc[((world_value['name']!='Kosovo') & (world_value['iso_a3']=='-99')), 'country_appear']= 0
    world_value = world_value.fillna(0)
    
    world_value = world_value.drop('country_name', axis =1)
    
    return country_frame, world_value
    
```

```python
def draw_country_frame(info_article_k):
    
    country_frame, world_value = get_country_frame(info_article_k)
    
    fig, ax = plt.subplots(1, figsize=(15,10))
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    plt.xticks([]) 
    plt.yticks([])  
    
    world_value.plot(column='country_appear',
                     ax = ax,
                     cmap='BuGn', 
                     scheme='UserDefined',
                     classification_kwds={'bins':[0,1,10,20,50]},
                     legend=True,
                     legend_kwds = {'loc': 'lower left',
                                    'title': 'Number of Articles',
                                    'shadow': False,
                                    'fmt': '{:.0f}',
                                    'labels': ['0','1','2-10','11-20','21-50','>50']}
                    )
    
    return country_frame
```

```python
def describe_judge(info_issue, type=1, lowess_frac=0.03):
    
    x = info_issue['period'].values
    y = info_issue['article_judge_share'].values
    y2 = info_issue['article_judge_quant'].values
    
    lowess=sm.nonparametric.lowess
    z=lowess(y,range(0,len(x)),frac=lowess_frac)
    z2=lowess(y2,range(0,len(x)),frac=lowess_frac)
    
    fig = plt.figure(figsize = (12,5))
    ax = fig.add_subplot(1,1,1)
    
    ax.set_xlabel('')
    
    # ax.set_xlim('1999-01-01','2021-01-01')
    ax.set_xticks(range(0,len(x),10))
    # ax.set_xticklabels() 
    
    if type == 1:
        ax.set_ylabel('Percentage of Articles with Multiple Countries', fontsize=10)
        ax.set_ylim(-0.3,1.3)
        ax.scatter(x, y, marker='o',s=9,
                   facecolors='none',edgecolors='#8c564b')
        ax.plot(z[:,0],z[:,1],color='#8c564b',lw=1.5)
        img = ax.get_figure()
        img.savefig(path+'\\picture\\'+'lowess_percentage_issue_'+str(lowess_frac)+'.png')
        
    elif type == 2:
        ax.set_ylabel('Number of Articles with Multiple Countries', fontsize=10)
        ax.set_ylim(-1,20)
        ax.scatter(x, y2, marker='o',s=9,
                   facecolors='none',edgecolors='#8c564b')
        ax.plot(z[:,0],z2[:,1],color='#8c564b',lw=1.5)
        img = ax.get_figure()
        img.savefig(path+'\\picture\\'+'lowess_number_issue_'+str(lowess_frac)+'.png')
        
    return img
```
