```python
def title_to_word(i_article):
    
    # porter_stemmer = PorterStemmer()
    # porter_stemmer.stem('presumably')
    # nltk.pos_tag(['institution'])
    wordnet_lematizer = WordNetLemmatizer()
    
    text = info_article.iloc[i_article]['name_book']
    
    words = nltk.word_tokenize(text)
    
    words = [wordnet_lematizer.lemmatize(i_word) for i_word in words]
    words = [i_word for i_word in words if i_word not in stopwords.words('english')]
    
    english_punctuations = [',','.',':',';','?','!','(',')','[','’',
                            ']','@','&','#','%','$','{','}','--','-']
    words = [i_word for i_word in words if i_word not in english_punctuations]
    
    return words
```

```python
for i_article in range(info_article.shape[0]):
    
    info_article.loc[i_article, 'name_book_words'] = str(title_to_word(i_article))
```

```python
# info_article.columns.values

# info_article['keyword']
keywords_list = info_article['keyword'].values

keywords = []
for i_keyword in keywords_list:
    i_keyword = eval(i_keyword)
    keywords = keywords+i_keyword
# print(keywords)

# info_article['name_book_words']
name_book_words_list = info_article['name_book_words']

name_book_words = []
for i_name_book_words in name_book_words_list:
    i_name_book_words = eval(i_name_book_words)
    for i_word in i_name_book_words:
        if (i_word == '‘') | (i_word == '–') | (i_word == "'") | (i_word == "'s"):
            continue
        else:
            name_book_words.append(i_word)      
# print(name_book_words)
```

```python
# part1 : wordcloud

# print(keywords)
wordcloud = WordCloud(background_color='white',
                      contour_color='steelblue',
                      max_words=5000,
                      contour_width=3,
                      width=600,
                      height=300)
# type(keywords)
# ','.join(keywords)
keyword_string = ''
for i_keyword in keywords:
    try:
        keyword_string = keyword_string + ',' + i_keyword
    except:
        continue

wordcloud.generate(keyword_string)
wordcloud.to_image()
```

```python
# part 2 : prepare data for LDA analysis
import gensim
import gensim.corpora as corpora

# keywords

keywords_list = info_article['keyword'].values
keywords_list_r = []
for i_keyword in keywords_list:
    
    i_keyword = eval(i_keyword)
    
    i_keyword_r = []
    for i_i_keyword in i_keyword:
        if i_i_keyword != None:
            i_keyword_r.append(i_i_keyword)
    
    keywords_list_r.append(i_keyword_r)
    
# name_book_words
```

```python
keywords_dict = corpora.Dictionary(keywords_list_r)
```

```python
keywords_dict
```

```python
print(keywords_dict)
```

```python
corpus = [keywords_dict.doc2bow(i_keyword) for i_keyword in keywords_list_r]
```

```python
corpus[:4]
```

```python
from pprint import pprint

lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=keywords_dict,
                                       num_topics=10,
                                       random_state=0)

pprint(lda_model.print_topics())
# doc_lda = lda_model[corpus]
```
