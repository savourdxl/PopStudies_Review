```python
def find_country(i_article):
    
    author_info = eval(info_article.iloc[i_article]['author_info'])
    
    if author_info == []:
        return []
    
    author_info_new = []
    
    for i_author_info in author_info:
        
        institution = i_author_info[2].strip(' ')
        
        if institution == '':
            country = ''
            continue
        
        # special case
        institution = institution.replace('University of York','United Kindom')
        institution = institution.replace('York University','Canada')
        
        # delete those words that are not important
        institution = institution.replace('University','')
        institution = institution.replace('Institute','')
        institution = institution.replace('School','')
        institution = institution.replace('Department','')
        institution = institution.replace('Dept','')
        institution = institution.replace('Collège','')
        institution = institution.replace('Centre','')
        institution = institution.replace('Universidad','')
        institution = institution.replace('Universität','')
        institution = institution.replace('Ecole','')
        institution = institution.replace('Social','')
        institution = institution.replace('Sociology','')
        institution = institution.replace('Research','')
        institution = institution.replace('Policy','')
        institution = institution.replace('Comparative','')
        institution = institution.replace('Analysis','')
        institution = institution.replace('College','')
        institution = institution.replace('The Woodlands','')
        institution = institution.replace('Sciences','')
        institution = institution.replace('Economic','')
        institution = institution.replace('Management','')
        institution = institution.replace('Universitad','')
        institution = institution.replace('C/Trinidad','')
        institution = institution.replace('Newcastle','')
        institution = institution.replace('National','')
        institution = institution.replace('Security','')
        institution = institution.replace('Justice','')
        institution = institution.replace('Global','')
        institution = institution.replace('J. Mitchell','')
        institution = institution.replace('Swiss','')
        institution = institution.replace('Federal','')
        institution = institution.replace('Bussiness','')
        institution = institution.replace('Marketing','')
        institution = institution.replace('Educational','')
        institution = institution.replace('Trajectories','')
        institution = institution.replace('Leibniz','')
        institution = institution.replace('Council','')
        institution = institution.replace('Policies','')
        
        # modification: country errors
        institution = institution.replace('UK','United Kingdom')
        institution = institution.replace('USA','United States')
        
        # modification: university name errors
        institution = institution.replace('Essex','United Kingdom')
        institution = institution.replace('St Andrews','United Kingdom')
        institution = institution.replace('Oxford','United Kingdom')
        institution = institution.replace('Cologne','Germany')
        institution = institution.replace('Kent','United Kingdom')
        institution = institution.replace('Lincoln','United Kingdom')
        institution = institution.replace('Surrey','United Kingdom')
        institution = institution.replace('Durham','United Kingdom')
        institution = institution.replace('Limerick','Ireland')
        institution = institution.replace('York','')
        institution = institution.replace('Warwick','United Kingdom')
        institution = institution.replace('Maximilians','Germany')
        institution = institution.replace('Kingston','United Kingdom')
        
        # modification: cities errors
        institution = institution.replace('London','United Kingdom')
        institution = institution.replace('Umea','Sweden')
        institution = institution.replace('Geneva','Switzerland')
        institution = institution.replace('Oviedo','Spain')
        institution = institution.replace('Canterbury','United Kingdom')
        institution = institution.replace('Petersburg','Russia')
        institution = institution.replace('Tel Avi','Israel')
        institution = institution.replace('Washington','United States')
        institution = institution.replace('Jaen','Spain')
        institution = institution.replace('Naples','Italy')
        institution = institution.replace('Giessen','Germany')
        institution = institution.replace('Bamberg','Germany,')
        
        places = GeoText(institution)
        country = list(places.country_mentions)
        
        country_name = []
        
        for i_country in country:
            if i_country == 'XK':
                name = 'Kosovo'
            else:
                name = pycountry.countries.get(alpha_2=i_country).name
            country_name.append(name)
        
        my_tuple = (i_author_info[0],i_author_info[1], str(country_name), 
                    i_author_info[2],i_author_info[3])
        
        author_info_new.append(my_tuple)
        
    return author_info_new
```

```python
for i_article in range(info_article.shape[0]):
    
    info_article.loc[i_article, 'author_info'] = str(find_country(i_article))
```
